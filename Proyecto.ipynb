# **Introducción**

La comuna de Olón cuenta con un sistema de abastecimiento de agua subterránea que depende de 14 pozos distribuidos en su acuífero dentro de una zona costera. Desde el año 2020, la Junta de Agua ha llevado registros manuales de parámetros hidráulicos y físico-químicos asociados al funcionamiento de estos pozos. Entre las variables registradas se incluyen nivel estático, nivel dinámico, conductividad eléctrica, pH, temperatura, salinidad, resistividad, sólidos disueltos totales (TDS) y oxígeno disuelto(DO).

La importancia de analizar esta información se relaciona con las condiciones propias de un acuífero costero, donde la calidad del agua puede variar por factores hidrogeológicos, climáticos y operativos. En Olón existen periodos donde el abastecimiento puede verse afectado por el estiaje, así como por el incremento de la demanda de agua durante temporadas turísticas. Estas condiciones hacen necesario estudiar el comportamiento de la salinidad y su posible relación con otras variables del sistema (Herrera-Franco et al., 2020; Javier Montalván et al., 2023).

La base de datos utilizada en este proyecto integra información de los 14 pozos durante una ventana de analisis entre el 2020 a 2025. Además de los registros manuales de campo, se incorporaron variables espaciales extraídas a escala de pozo mediante herramientas como Google Earth Engine y ArcGIS (Amani et al., 2020; Hoyer & Hamman, 2017). Estas variables provienen de fuentes geoespaciales como modelos digitales de elevación y capas ambientales, permitiendo incluir información sobre elevación, pendiente, acumulación de flujo, densidad de drenaje, índice topográfico de humedad (TWI), distancia a la línea costera, distancia al río Olón y el Índice de Vegetación de Diferencia Normalizada (NVDI) (Adeyemi & Komolafe, 2025).

El presente proyecto se enfoca en realizar un análisis exploratorio de datos, conocido como EDA, con el propósito de revisar la estructura, calidad y comportamiento general de la base. Este análisis permitirá identificar valores faltantes, inconsistencias, distribución de variables, diferencias entre pozos, tendencias temporales y posibles valores atípicos(outliers). Aunque en una etapa futura se busca desarrollar un modelo predictivo de salinidad, esta primera fase se centra únicamente en la limpieza, exploración y visualización de los datos disponibles.

## **1. Importar librerías y carga de datos** 
Se importaron las librerías necesarias para todo el análisis y visualización de datos. Posteriormente, se cargó la base de datos de los pozos de Olón en un DataFrame llamado 'data_pozos.csv', con el fin de iniciar la revisión exploratoria de la información.

## **2. Revisión inicial de la base de datos**
Se realizó una revisión inicial de la base de datos para conocer su estructura general. En esta etapa se observaron las primeras filas, los nombres de las columnas y los tipos de datos de cada variable.

```python

# Revisión de nombres de columnas
data_pozos.columns
# Información general de la base: columnas, datos no nulos y tipos de datos
data_pozos.info()
# Revisión específica de los tipos de datos
data_pozos.dtypes
# Visualización de las primeras filas
data_pozos.head()
```

## **3. Normalización de nombres de columnas**
Se creó una copia de la base original para evitar modificar directamente el DataFrame inicial. Luego, se normalizaron los nombres de las columnas, transformándolos a minúsculas, eliminando espacios innecesarios y reemplazando espacios internos por guiones bajos. Además, la columna año fue renombrada como anio para facilitar su uso en Python.

```python
# Creación de una copia de trabajo
data_limpia = data_pozos.copy()

# Revisión de nombres originales
print("Columnas originales:")
print(data_limpia.columns.tolist())

# Normalización de nombres de columnas
data_limpia.columns = (
    data_limpia.columns
    .str.strip()
    .str.lower()
    .str.replace(" ", "_", regex=False)
)

# Renombrar columna con carácter especial
data_limpia = data_limpia.rename(columns={"año": "anio"})

# Revisión de nombres normalizados
print("Columnas normalizadas:")
print(data_limpia.columns.tolist())
```

## **4. Clasificación de variables**
Se clasificaron las variables de la base de datos según su función dentro del análisis. Esta organización permitió separar variables de identificación, ubicación, características espaciales, condiciones temporales y climáticas, variables hidráulicas y parámetros físico-químicos.

```python
# Clasificación de variables según su función dentro del análisis

variables_identificacion = ['id', 'estado']
variables_ubicacion = ['x', 'y']
variables_pozo = ['brocal', 'elev', 'prof_pozo']
variables_espaciales = ['pendiente', 'flow_ac', 'densidad_drenaje','twi', 'dist_cost', 'dist_rio_m', 'ndvi']
variables_temporales_climaticas = [ 'anio', 'mes', 'lluviamensual_mm', 'et_mensual']
variables_hidraulicas = ['nivel_est', 'nivel_din']
variables_fisico_quimicas = ['ce', 'ph', 'temperatura', 'salinidad', 'resistividad', 'tds', 'do']

# Agrupación de todas las variables del proyecto
grupos_variables = {
    'Identificación': variables_identificacion,
    'Ubicación': variables_ubicacion,
    'Características del pozo': variables_pozo,
    'Espaciales': variables_espaciales,
    'Temporales y climáticas': variables_temporales_climaticas,
    'Hidráulicas': variables_hidraulicas,
    'Físico-químicas': variables_fisico_quimicas}

# Revisión de variables por grupo
for grupo, variables in grupos_variables.items():
    print(f"{grupo}: {variables}")
```

---

## **5. Conversión de tipos de datos**
Se revisaron y ajustaron los tipos de datos de las variables numéricas. Debido a que la base proviene parcialmente de registros manuales, algunas columnas podían estar almacenadas como texto o contener separadores decimales con coma, por lo que fueron transformadas a formato numérico para permitir cálculos estadísticos y visualizaciones. En este bloque, si existían valores como --, -, espacios vacíos o textos dentro de columnas numéricas, se convertirán automáticamente en NaN.

```python
# Definición de variables numéricas del proyecto
columnas_numericas = ( variables_ubicacion +variables_pozo +variables_espaciales +variables_temporales_climaticas +variables_hidraulicas +variables_fisico_quimicas)

# Conversión de columnas numéricas almacenadas como texto
for col in columnas_numericas:
    data_limpia[col] = (
        data_limpia[col]
        .astype(str)
        .str.replace(",", ".", regex=False)
        .str.strip()
    )
    data_limpia[col] = pd.to_numeric(data_limpia[col], errors="coerce")

# Revisión de tipos de datos después de la conversión
data_limpia[columnas_numericas].dtypes.reset_index()
```

---

## **6. Revisión de valores nulos**
Se revisaron los valores no válidos generados durante la conversión de datos numéricos. Esta revisión permitió identificar columnas donde existían registros vacíos, símbolos o textos que no podían ser interpretados como números por Python.

```python
# Revisión de valores nulos generados o presentes después de la conversión numérica
nulos_por_columna = data_limpia[columnas_numericas].isnull().sum()

# Filtrar únicamente columnas con valores nulos
nulos_por_columna[nulos_por_columna > 0].sort_values(ascending=False)
# Porcentaje de valores nulos por columna numérica
porcentaje_nulos = (
    data_limpia[columnas_numericas]
    .isnull()
    .mean()
    .mul(100)
    .round(2)
)

porcentaje_nulos[porcentaje_nulos > 0].sort_values(ascending=False)
# Resumen de valores nulos por columna numérica
resumen_nulos = pd.DataFrame({
    "nulos": nulos_por_columna,
    "porcentaje": porcentaje_nulos})

resumen_nulos = resumen_nulos[resumen_nulos["nulos"] > 0].sort_values(by="porcentaje",ascending=False).reset_index()

resumen_nulos
```

## **8. Revisión de categorías y creación de variables auxiliares**
Se revisaron las variables categóricas principales de la base de datos, especialmente el identificador del pozo y el estado operativo. Además, se creo una variables auxiliar para facilitar la interpretación del análisis:categoria_salinidad, según los rangos establecidos en el TULSMA / Libro VI, Anexo 1. Esta clasificación permitió diferenciar los registros en agua dulce, agua salobre y agua salina.

```python
# Revisión de categorías de la variable estado
data_limpia['estado'].value_counts()
# Normalización básica de la variable estado
data_limpia['estado'] = (
    data_limpia['estado']
    .astype('string')
    .str.strip()
    .str.lower()
)

data_limpia['estado'].value_counts()
# Revisión del número de registros por pozo
data_limpia['id'].value_counts().sort_index()

# Clasificación de salinidad según rangos TULSMA / Libro VI, Anexo 1
condiciones_salinidad = [
    data_limpia['salinidad'] <= 0.5,
    (data_limpia['salinidad'] > 0.5) & (data_limpia['salinidad'] < 30),
    data_limpia['salinidad'] >= 30
]

categorias_salinidad = [
    'agua dulce',
    'agua salobre',
    'agua salina'
]

# Se usa "sin dato" como valor temporal para evitar conflicto de tipos
data_limpia['categoria_salinidad'] = np.select(
    condiciones_salinidad,
    categorias_salinidad,
    default='sin dato'
)

# Se reemplaza "sin dato" por NaN para mantener los valores faltantes correctamente
data_limpia['categoria_salinidad'] = data_limpia['categoria_salinidad'].replace('sin dato', np.nan)

# Revisión de la variable categórica creada
data_limpia['categoria_salinidad'].value_counts(dropna=False)

# Conversión de variables categóricas a tipo category
data_limpia['id'] = data_limpia['id'].astype('category')
data_limpia['estado'] = data_limpia['estado'].astype('category')
data_limpia['categoria_salinidad'] = data_limpia['categoria_salinidad'].astype('category')

data_limpia[['id', 'estado', 'categoria_salinidad']].dtypes
```

Se identificaron 85 registros con inconsistencias entre el estado operativo del pozo y la disponibilidad del nivel dinámico. Para mantener la trazabilidad, no se modificó la variable original estado; en su lugar, se creó la variable estado_corregido.

```python
# Se crea una copia para aplicar correcciones sin modificar data_limpia
data_corregida = data_limpia.copy()

# Se crea una nueva columna para conservar el estado original y registrar el estado corregido
data_corregida['estado_corregido'] = data_corregida['estado']

# Caso 1: operativo sin nivel dinámico se reclasifica como no operativo
data_corregida.loc[
    (data_corregida['estado'] == 'operativo') &
    (data_corregida['nivel_din'].isnull()),
    'estado_corregido'
] = 'no operativo'

# Caso 2: no operativo con nivel dinámico se reclasifica como operativo
data_corregida.loc[
    (data_corregida['estado'] == 'no operativo') &
    (data_corregida['nivel_din'].notnull()),
    'estado_corregido'
] = 'operativo'


# Comparación entre estado original y estado corregido
data_corregida[['estado', 'estado_corregido']].value_counts()
```

## 12. Criterios para el tratamiento de valores nulos
Se definieron criterios para el tratamiento de valores nulos a partir del porcentaje de datos faltantes por variable y por pozo. Esta revisión permitió diferenciar variables completas, variables con pocos datos faltantes, variables que requieren revisión y casos donde no sería recomendable imputar debido a ausencia total de información en un pozo.

```python
# Tabla de criterios generales para el tratamiento de valores nulos
criterios_tratamiento = pd.DataFrame({
    'porcentaje_nulos': [
        '0%',
        '> 0% y <= 20%',
        '> 20% y < 100%',
        '100%'
    ],
    'criterio': [
        'Variable completa, no requiere tratamiento.',
        'Imputación posible, según el tipo de variable.',
        'Revisar antes de imputar, porque existe ausencia importante de datos.',
        'No imputar con promedio del pozo, porque no existe información real para esa variable.'
    ]
})

criterios_tratamiento
```

```python
# Creación de una tabla de criterios por pozo y variable
criterios_nulos_pozo_variable = pd.DataFrame(
    index=tabla_nulos_pozo_variable.index,
    columns=tabla_nulos_pozo_variable.columns
)

criterios_nulos_pozo_variable[tabla_nulos_pozo_variable == 0] = 'completo'
criterios_nulos_pozo_variable[
    (tabla_nulos_pozo_variable > 0) & (tabla_nulos_pozo_variable <= 20)
] = 'imputación posible'

criterios_nulos_pozo_variable[
    (tabla_nulos_pozo_variable > 20) & (tabla_nulos_pozo_variable < 100)
] = 'revisar antes de imputar'

criterios_nulos_pozo_variable[tabla_nulos_pozo_variable == 100] = 'no imputar'

criterios_nulos_pozo_variable
```

## **13. Tratamiento inicial de valores nulos**
Se aplicó un tratamiento inicial de valores nulos considerando el porcentaje de datos faltantes por variable y por pozo. Para los casos con hasta 20% de nulos, se realizó una imputación con el promedio de la misma variable dentro del mismo pozo. Los casos con mayor porcentaje de ausencia fueron separados para revisión, mientras que los casos con 100% de nulos no fueron imputados.

```python
# Creación de una copia para aplicar el tratamiento de nulos
data_tratada = data_corregida.copy()
# Conversión temporal del ID para ordenar los pozos numéricamente
id_pozo = pd.to_numeric(data_tratada['id'].astype(str), errors='coerce')

# Variables principales con nulos
nulos_por_variable = data_tratada[variables_nulos].isnull().sum()
variables_con_nulos = nulos_por_variable[nulos_por_variable > 0].index.tolist()

# Porcentaje de nulos por variable en cada pozo
porcentaje_nulos_pozo_variable = (
    data_tratada[variables_nulos]
    .isnull()
    .groupby(id_pozo)
    .mean()
    .mul(100)
    .round(2)
)

# Mostrar solo variables con nulos
tabla_nulos_pozo_variable = porcentaje_nulos_pozo_variable[variables_con_nulos]

tabla_nulos_pozo_variable

# Casos con imputación posible: más de 0% y hasta 20% de nulos
tabla_imputacion_posible = tabla_nulos_pozo_variable[
    (tabla_nulos_pozo_variable > 0) & 
    (tabla_nulos_pozo_variable <= 20)
]

tabla_imputacion_posible.dropna(how='all')
```

```python
# Casos que requieren revisión antes de imputar: más de 20% y menos de 100%
tabla_revisar_imputacion = tabla_nulos_pozo_variable[
    (tabla_nulos_pozo_variable > 20) & 
    (tabla_nulos_pozo_variable < 100)
]

tabla_revisar_imputacion.dropna(how='all')
```

```python
# Casos con 100% de nulos
tabla_no_imputar = tabla_nulos_pozo_variable[
    tabla_nulos_pozo_variable == 100
]

tabla_no_imputar.dropna(how='all')
```

```python
# Variables que no se imputan automáticamente en esta etapa
variables_excluir_imputacion = ['salinidad', 'do', 'nivel_din']

# Variables que sí pueden evaluarse para imputación
variables_para_imputar = [
    col for col in tabla_imputacion_posible.columns
    if col in columnas_numericas and col not in variables_excluir_imputacion
]

variables_para_imputar
```

```python
# Imputación con el promedio de cada variable dentro de cada pozo
for pozo in tabla_imputacion_posible.index:
    for variable in variables_para_imputar:
        
        porcentaje_nulo = tabla_imputacion_posible.loc[pozo, variable]
        
        if pd.notnull(porcentaje_nulo):
            promedio_pozo = data_tratada.loc[id_pozo == pozo, variable].mean()
            
            data_tratada.loc[
                (id_pozo == pozo) & (data_tratada[variable].isnull()),
                variable
            ] = promedio_pozo
```

```python
# Imputación con el promedio de cada variable dentro de cada pozo
for pozo in tabla_imputacion_posible.index:
    for variable in variables_para_imputar:
        
        porcentaje_nulo = tabla_imputacion_posible.loc[pozo, variable]
        
        if pd.notnull(porcentaje_nulo):
            promedio_pozo = data_tratada.loc[id_pozo == pozo, variable].mean()
            
            data_tratada.loc[
                (id_pozo == pozo) & (data_tratada[variable].isnull()),
                variable
            ] = promedio_pozo
```

```python
# Revisión de nulos restantes por variable
nulos_restantes = data_tratada[variables_nulos].isnull().sum()

nulos_restantes[nulos_restantes > 0].sort_values(ascending=False)
```

```python
# Comparación de nulos antes y después del tratamiento
nulos_antes = data_limpia[variables_nulos].isnull().sum().sum()
nulos_despues = data_tratada[variables_nulos].isnull().sum().sum()

print(f"Nulos antes del tratamiento: {nulos_antes}")
print(f"Nulos después del tratamiento: {nulos_despues}")
print(f"Nulos imputados: {nulos_antes - nulos_despues}")
```

# **Conclusiones y recomendaciones**

## **Conclusiones**

El análisis exploratorio permitió identificar que la salinidad presentó una variación importante dentro del conjunto de datos. Se analizaron 635 registros válidos de salinidad, con 27 valores nulos. **El valor mínimo fue de 0.03 UPS y el máximo de 5.00 UPS, lo que representó una variación total de 4.97 UPS. La media general fue de 0.846 UPS y la mediana de 0.50 UPS, lo cual indicó que algunos valores altos influyeron en el promedio general.**

La salinidad no se distribuyó de manera uniforme entre los pozos. **Los pozos con mayores valores promedio fueron el ID 4, con 2.615 UPS, y el ID 5, con 2.136 UPS. En contraste, el pozo ID 12 presentó el menor promedio, con 0.427 UPS. La diferencia entre el pozo con mayor y menor salinidad promedio fue de 2.188 UPS, lo que evidenció comportamientos hidroquímicos distintos entre puntos de monitoreo. Además, el pozo ID 11 presentó un valor máximo de 5.00 UPS, aunque su promedio fue bajo, por lo que este caso debe revisarse como posible evento puntual o valor extremo.**

A nivel temporal, **el año con mayor variación de salinidad fue 2022, con una diferencia de 4.744 UPS entre el valor mínimo y máximo**. La media anual mostró una disminución desde 2020 hasta 2024, pasando de 1.327 UPS a 0.720 UPS, aunque el año 2020 debe interpretarse con precaución porque tuvo solo 22 registros. En 2025 se observó un ligero aumento de la media a 0.795 UPS. A nivel mensual, febrero y octubre presentaron los mayores promedios de salinidad, ambos con 0.928 UPS, mientras que junio presentó el menor promedio, con 0.706 UPS.

Según la clasificación aplicada, se identificaron 331 registros de agua dulce y 304 registros de agua salobre. No se registraron aguas salinas bajo el criterio de clasificacion de la TULSMA. **Los pozos ID 4, ID 5, ID 13 y ID 14 presentaron predominio de registros salobres, mientras que los pozos ID 12, ID 1, ID 9, ID 7 e ID 8 presentaron mayor cantidad de registros de agua dulce. Esto permitió diferenciar pozos con condiciones de mayor influencia salina frente a pozos con condiciones de menor salinidad.**

El análisis de correlación mostró una **relación positiva muy fuerte entre salinidad y conductividad eléctrica, con un coeficiente de 0.944**. También se observó una relación positiva entre salinidad y TDS, con 0.610, y una relación negativa entre salinidad y resistividad, con -0.359. Estas relaciones fueron coherentes con el comportamiento hidrogeoquímico esperado, ya que el aumento de sales disueltas incrementa la conductividad eléctrica y reduce la resistividad. Además, la alta relación entre salinidad, CE y TDS evidenció posible colinealidad, por lo que estas variables deberán revisarse cuidadosamente si se desarrolla un modelo predictivo en una etapa posterior.

En cuanto al tratamiento de valores nulos, **se identificaron inicialmente 753 datos faltantes dentro de las variables principales. Después del tratamiento, quedaron 581 valores nulos, por lo que se imputaron 172 registros. La imputación se aplicó únicamente cuando existía información suficiente dentro del mismo pozo, evitando reemplazos generales que pudieran alterar la interpretación hidrogeológica**. **No se imputaron los valores de DO porque esta variable empezó a medirse desde 2024, por lo que sus nulos corresponden a ausencia histórica de medición. Tampoco se imputó la salinidad, debido a que es la variable principal del análisis. En el caso de nivel dinámico, se conservaron 180 valores nulos asociados a pozos no operativos, ya que no existió bombeo y, por tanto, no se registró nivel dinámico**. Además, se identificaron 85 inconsistencias entre estado operativo y disponibilidad de nivel dinámico, por lo que se creó una variable auxiliar de estado corregido para mantener la trazabilidad sin modificar directamente la variable original.

En cuanto a la completitud de la base, se identificaron diferencias en la cantidad de registros por pozo. Los pozos ID 1, 2, 4, 5, 6, 7, 8, 9 y 10 tuvieron 50 registros cada uno, mientras que los pozos ID 3 e ID 14 tuvieron la menor cantidad, con 39 registros. Esta diferencia debe considerarse al comparar resultados entre pozos, ya que no todos cuentan con la misma cantidad de información disponible. En conjunto, el análisis permitió reconocer patrones espaciales, temporales e hidroquímicos de la salinidad, así como documentar las decisiones de limpieza necesarias para conservar la calidad e interpretación de los datos.

## **Recomendaciones**

Se recomienda mantener un registro más uniforme de las mediciones por pozo, año y mes, con el fin de reducir vacíos de información y mejorar la comparación temporal entre pozos.

Se recomienda registrar de forma clara la condición operativa de cada pozo durante la medición, especialmente cuando no sea posible encender la bomba. Esto permitirá diferenciar entre datos faltantes por error de registro y datos ausentes por condiciones reales de operación.

Se recomienda continuar midiendo DO en los próximos años para fortalecer el análisis físico-químico del agua subterránea, ya que esta variable solo cuenta con registros desde 2024.

Se recomienda no eliminar automáticamente los valores atípicos observados en salinidad, conductividad eléctrica o TDS, sino revisarlos con criterio hidrogeológico, debido a que pueden representar condiciones reales del acuífero costero.

Se recomienda complementar el análisis exploratorio con una futura etapa de modelación predictiva de salinidad, utilizando únicamente una base previamente depurada y considerando variables espaciales, climáticas, hidráulicas y físico-químicas.
